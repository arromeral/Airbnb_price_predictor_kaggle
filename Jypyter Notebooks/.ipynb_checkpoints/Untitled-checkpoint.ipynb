```python
plt.figure(figsize=(15, 10))   # 150x100 pixeles

sns.set(style='white')    # estilo blanco

mask=np.triu(np.ones_like(bnb2.corr(), dtype=bool))    # mascara para tapar lo de arriba

cmap=sns.diverging_palette(0, 10, as_cmap=True)   # paleta de  colores


sns.heatmap(bnb2.corr(),    
           mask=mask,
          cmap=cmap,
          center=0,
          square=True,
          annot=True,
          linewidths=0.5,
          cbar_kws={'shrink': 0.5});
```

```python
bnb2 = pd.get_dummies(bnb2, columns=['property_type','room_type','bed_type','cancellation_policy','city','host_experience'], drop_first=True)
```

```python
X = bnb2.drop('log_price', axis=1)

y = bnb2.log_price
```

```python

X_train, X_test, y_train, y_test = tts(X, y, 
                                       train_size=0.8, 
                                       test_size=0.2,
                                       random_state=42)


X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
escalador = StandardScaler()    # inicia el normalizador
 
escalador.fit(X_train[['accommodates', 'bathrooms','bedrooms', 'bathrooms']])   # ajusta segun train, saca media y std
 
X_train[['accommodates', 'bathrooms','bedrooms', 'bathrooms']] = escalador.transform(X_train[['accommodates', 'bathrooms','bedrooms', 'bathrooms']])   # transforma train
```

```python
X_test[['accommodates', 'bathrooms','bedrooms', 'bathrooms']] = escalador.transform(X_test[['accommodates', 'bathrooms','bedrooms', 'bathrooms']])  # transforma test
```

```python
X_test.head()
```

```python
X_test.head()
```

```python
rfr = RFR()    # inicia el modelo


rfr.fit(X_train, y_train)   # entrena el modelo


y_pred = rfr.predict(X_test)   # prediccion con el tamaÃ±o del test
```

```python
mse(y_test, y_pred, squared=False)
```
